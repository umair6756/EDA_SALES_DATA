```python
def outliers_graphs(x):
    sns.boxplot(x=df[x])
    plt.title(f"Outliers of {x}")
    plt.show()

outliers_graphs("Profit")
outliers_graphs("Sales")
```

```python
def bar_chart(x):
   category_profit = df.groupby(x)["Profit"].sum().reset_index()
   print(category_profit)
   plt.pie(category_profit["Profit"],labels=category_profit[x],startangle=90,autopct='%1.1f%%')
   plt.title(f"Profit by {x}")
   plt.legend()
   plt.show()

bar_chart("Category")
bar_chart("Region")
bar_chart("Country")

```

```python
reign_profit = df.groupby("Region")["Profit"].sum().reset_index()
print(reign_profit)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Year"] = df["Order Date"].dt.year.astype(str) 
```

```python
def bar_chart(x):
   year_profit = df.groupby("Year")[x].sum().reset_index()
   print(year_profit)
   plt.bar(year_profit["Year"],year_profit[x], edgecolor='black')
   plt.title(f"{x} per year")
   plt.xlabel("Year")
   plt.ylabel(x)
   for y,p in enumerate(year_profit[x]):
      plt.text(x=y,y=p + 1000,s=f'{p}',ha="center",fontsize=10)
   plt.show()

bar_chart("Profit")
bar_chart("Quantity")
```
